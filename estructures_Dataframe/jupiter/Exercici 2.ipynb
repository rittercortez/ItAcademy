- Taula de les aerolínies amb més endarreriments acumulats

```python
import pandas as pd

df = pd.read_csv("DelayedFlights.csv")

# Resumeix estadísticament les columnes d’interès
filtrado = df[["Year","Month","DayofMonth","DayOfWeek","Origin","Dest","Diverted","Cancelled",
               "CancellationCode","WeatherDelay","LateAircraftDelay"]]


"""
Troba quantes dades faltants hi ha per columna

- la función "isna()" nos permite buscar elementos nulos/faltantes
- la función "sum()" nos permite sumar todos los elementos faltantes
"""
df.isna().sum()


"""
Crea columnes noves (velocitat mitjana del vol..
Creo una funcion a la cual le envio los datos de cada columna para calcular la Velocidad Media 

"""
def velocidad_media(velocidad):
    resultado = velocidad["CRSElapsedTime"]-velocidad["ActualElapsedTime"]/velocidad["Distance"]
    return resultado

df["Velocidad Media"] = df.apply(velocidad_media, axis = 1)
df["Velocidad Media"]

"""
si ha arribat tard o no...)
la funcion "notna()" nos permite invertir los resultado de los objetos faltantes
"""
def retraso(delay):
    if delay["ArrDelay"] > 0:
        x = "True"
        return x
    else:
        y = "False"
        return y

df["Retraso (True / False)"] = df.apply(retraso, axis = 1)
df["Retraso (True / False)"]


"""
Taula de les aerolínies amb més endarreriments acumulats
- con la funcion "agg()" me permito agrupar los nombres de columnas para sumarlo
"""
df.groupby("UniqueCarrier").agg({
    "ArrDelay" : "sum"
})

# Quins són els vols més llargs? I els més endarrerits?
# Los vuelos mas Largos 
# La función "sort_values" nos permite ordenar valores y el parametro
# "ascending=False" nos indica que lo filtraremos de forma descendente
largos = df[["Origin", "Dest", "Distance"]]
largos.sort_values("Distance", ascending = False)

# Los Vuelos  con mas retraso
retraso = df[["Origin", "Dest", "ArrDelay"]]
retraso.sort_values("ArrDelay", ascending = False)
```
